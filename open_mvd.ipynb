```
import pandas as pd 
import numpy as np
import re
mvd = pd.read_csv("russian_mvd.csv") 
mvd = pd.DataFrame(data=mvd)
print(mvd)
```

```
#drop NANs
mvd.dropna()
mvd.info(verbose=True)
mvd = mvd.drop(mvd.columns[[6,7,8,9,10,11]], axis=1) 
print(mvd)
```

```
mvd['adress'] = mvd['adress'].astype(str)
type(mvd['adress'])
```

```
import pandas as pd
import re as re
def find_phone_number(text):
    ph_no = re.findall(r": [+0-9-\s]+;",text)
    return "".join(ph_no)
mvd['number']=mvd['adress'].apply(lambda x: find_phone_number(x))
print("\Extracting numbers from dataframe columns:")
print(mvd[['number']])
```

```
def drop_trash(text):
  aa = re.sub(r';', '', text)
  bb = re.sub(r':', '', aa)
  return bb
mvd['phone']=mvd['number'].apply(lambda x: drop_trash(x))
print("\Extracting numbers from dataframe columns:")
print(mvd[['phone']])
```

```
print(mvd)
```

```
def find_email(text):
    ph_no = re.findall(r"[\w\.-]+@[\w\.-]+",text)
    return "".join(ph_no)
mvd['email']=mvd['adress'].apply(lambda x: find_email(x))
print("\Extracting numbers from dataframe columns:")
print(mvd[['email']])
```

```
def find_address(text):
    ph_no = re.findall(r"^[^;]*",text)
    return "".join(ph_no)
mvd['address']=mvd['adress'].apply(lambda x: find_address(x))
print("\Extracting numbers from dataframe columns:")
print(mvd[['address']])
```
